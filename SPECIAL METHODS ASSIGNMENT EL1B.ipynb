## Introduction
In Python, **Special Methods** (also known as **Dunder Methods** because they begin and end with double underscores, like `__init__`) are built-in methods that allow us to define how our custom objects behave when used with standard Python operations.

Instead of calling methods manually, special methods allow us to use built-in Python functions and operators (like `print(book)` or `len(book)`) directly on our objects.

```python
class SmartPhone:
    # 1. The __init__ method initializes the object's attributes
    def __init__(self, brand, model, storage_gb):
        self.brand = brand
        self.model = model
        self.storage_gb = storage_gb

    # 2. The __str__ method defines what happens when we print the object
    def __str__(self):
        return f"{self.brand} {self.model}"

    # 3. The __len__ method defines what len() returns (e.g., storage capacity)
    def __len__(self):
        return self.storage_gb

# --- Testing our Special Methods ---

# Creating an instance (Triggers __init__)
my_phone = SmartPhone("Apple", "iPhone 15", 128)

# Printing the object (Triggers __str__)
print(f"Phone details: {my_phone}")

# Checking the 'length' (Triggers __len__)
print(f"Storage capacity: {len(my_phosne)} GB")
```
